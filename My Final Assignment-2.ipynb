```python
# Define Foursquare credentials and version
CLIENT_ID = 'XXXXXXXXXXXXXE' 
CLIENT_SECRET = 'XXXXXXXXXXXX0' 
VERSION = '20200708'
```

```python
# Explore Neighborhoods in Kirkland
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng)
            
        results = requests.get(url).json()['response']['groups'][0]['items']
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
kirkland_venues = getNearbyVenues(names=df['Neighborhood'],
                                   latitudes=df['Latitude'],
                                   longitudes=df['Longitude']
                                  )
```

```python
print(kirkland_venues.shape)
kirkland_venues.head()
```

```python
print('There are {} uniques categories.'.format(len(kirkland_venues['Venue Category'].unique())))
```

```python
# Analyze each neighborhood
kirkland_onehot = pd.get_dummies(kirkland_venues[['Venue Category']], prefix="", prefix_sep="")

kirkland_onehot['Neighborhood'] = kirkland_venues['Neighborhood'] 

fixed_columns = [kirkland_onehot.columns[-1]] + list(kirkland_onehot.columns[:-1])
kirkland_onehot = kirkland_onehot[fixed_columns]

kirkland_onehot.head()
```

```python
kirkland_onehot.shape
```

```python
# Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category
kirkland_grouped = kirkland_onehot.groupby('Neighborhood').mean().reset_index()
kirkland_grouped
```

```python
kirkland_grouped.shape
```

```python
# Print each neighborhood along with the top 5 most common venues

num_top_venues = 5

for hood in kirkland_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = kirkland_grouped[kirkland_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# create a dataframe with above data
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = kirkland_grouped['Neighborhood']

for ind in np.arange(kirkland_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(kirkland_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

```python
# since there are only 11 neighborhoods have venue information pulled from Foursquare, before merging the venue information with the 
# previous dataframe, i will need to remove corresponding neighborhoods from previous dataframe
df.drop(df.index[[6,7]], inplace=True)
df
```

```python
# cluster neighborhoods
kclusters = 3

kirkland_grouped_clustering = kirkland_grouped.drop('Neighborhood', 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(kirkland_grouped_clustering)

kmeans.labels_[0:10]
```

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
kirkland_merged = df

kirkland_merged = kirkland_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

kirkland_merged
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(kirkland_merged['Latitude'], kirkland_merged['Longitude'], kirkland_merged['Neighborhood'], kirkland_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
